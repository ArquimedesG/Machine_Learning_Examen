## Parte 1: Regresión

```python
#  a. Importacion de librerias

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import subplots
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
#from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression

#  b. Asignacion de datos leidos del archivo csv a un dataframe llamado data
data= raw

# c.  Separacion de caracteristicas (X) y variable objetivo (y)
# Preprocesamiento
X = data.drop(columns=["clase_salario", "valor_salario"])
y = data["valor_salario"]

# TODO: Dividir los datos en conjuntos de entrenamiento y prueba

# Modelo de regresión
reg = LinearRegression()

#  a. Division de el conjunto de datos en conjuntos de entrenamiento y prueba (80%-20%).
X_train,X_test,y_train,y_test= train_test_split(X,y)

#  a. Utilizacion una regresión lineal para entrenar un modelo con el conjunto de entrenamiento.
reg.fit(X_train, y_train)

#  Obteniendo las predicciones del modelo usando el conjunto de prueba
y_pred = reg.predict(X_test)

# b. Evaluacion del rendimiento del modelo en el conjunto de prueba usando el error cuadrático medio (MSE).
# Evaluación
mse = mean_squared_error(y_test, y_pred)

print(f"Error Cuadrático Medio (sin estandarización): {mse}")
y_test_se=y_test.array
y_pred_se=y_pred

# TODO: Aplicar estandarización a las características y entrenar y evaluar el modelo de regresión nuevamente

# Using the StandardScaler function to normalize the data and assigning it to "X_std"
X_std = preprocessing.StandardScaler().fit_transform(X)

#  La variable objetivo (y) no se estandariza para poder realizar una comparación en el error cuadrático medio (MSE)
y=pd.DataFrame(y)
y_std=y

# Modelo de regresión
reg = LinearRegression()
X_train,X_test,y_train,y_test= train_test_split(X_std,y_std)
reg.fit(X_train, y_train)
y_pred = reg.predict(X_test)

#  Calculando coeficientes del modelo de regresión estandarizado
coef=reg.coef_

# Evaluación
mse = mean_squared_error(y_test, y_pred)
print()
print(f"Error Cuadrático Medio (con estandarización): {mse}")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.pipeline import Pipeline
import joblib


# Cargamos los datos
data = raw
X = data.drop(columns=["clase_salario", "valor_salario"])
y_reg = data["valor_salario"]
y_class = data["clase_salario"]

# Dividimos los datos
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X, y_reg, test_size=0.2, random_state=42)
X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(X, y_class, test_size=0.2, random_state=42)

# Creando el pipeline para regresión
pipeline_reg = Pipeline([
    ('scaler', StandardScaler()),
    ('regressor', LinearRegression())
])
pipeline_reg.fit(X_train_reg, y_train_reg)

# Creando el pipeline para clasificación
pipeline_class = Pipeline([
    ('scaler', StandardScaler()),
    ('classifier', LogisticRegression(max_iter=10000))
])
pipeline_class.fit(X_train_class, y_train_class)

# Exportando los modelos
joblib.dump(pipeline_reg, 'modelo_regresion.pkl')
joblib.dump(pipeline_class, 'modelo_clasificacion.pkl')
pipeline_class
```

```python
#  Usando el Modelo de Regresión creado con el Pipeline pipeline_reg

y_pred_reg = pipeline_reg.predict(X_test_reg)
print()

# Evaluación
mse = mean_squared_error(y_test_reg, y_pred_reg)
print()
print(f"Error Cuadrático Medio (con estandarización): {mse}")
```

```python
#  Usando el Modelo de Clasificación creado con el Pipeline pipeline_class

y_pred_class = pipeline_class.predict(X_test_class)
print()
print("Reporte de Clasificación (Modelo de Regresión Logística):")
print(classification_report(y_test_class, y_pred_class))
```
